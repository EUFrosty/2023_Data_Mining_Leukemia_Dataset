```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans
from scipy.cluster.hierarchy import dendrogram
from sklearn.metrics import silhouette_score, homogeneity_score
import matplotlib.cm as cm
from sklearn.metrics import silhouette_samples
from mpl_toolkits.mplot3d import Axes3D
from sklearn.preprocessing import StandardScaler
import plotly.express as px
```

```python
data = pd.read_csv('Leukemia.csv')
```

```python
X = data.drop('type', axis = "columns")
Y = data['type']
feature_names = X.columns
```

### Normalizacija podataka

```python
scaler = MinMaxScaler()
X = pd.DataFrame(scaler.fit_transform(X), columns = feature_names)
```

```python
sse = []
silh = []

ks = [i for i in range(5,20)]
for k in range(5,20):
    kmeans = KMeans(n_clusters = k, n_init = 'auto')
    kmeans.fit(X)
    sse.append(kmeans.inertia_)
    silh.append(silhouette_score(X, kmeans.labels_))
```

```python
plt.plot(ks, sse, marker = 'o')
plt.xlabel("Number of clusters")
plt.ylabel("SSE")
```

```python
plt.plot(ks, silh, marker = 'o')
plt.xlabel("Number of clusters")
plt.ylabel("Silhouette")
```

Pomocu pravila lakta i grafika promene siluete vidimo da je otprilike podjednako dobro uzeti da broj klastera bude 9 ili 13

```python
km9 = KMeans(n_clusters = 10, n_init = 'auto')
km13 = KMeans(n_clusters = 14, n_init = 'auto')
km9.fit(X)
km13.fit(X)
```

## PCA vizuelizacija izbora broja klastera

```python
ks = [i for i in range(2, 17)]
sse = []
silh = []
fig, axs = plt.subplots(5, 3, figsize=(10, 20))
fig.suptitle("")

for k, ax in zip(ks, axs.flatten()):
    kmeans = KMeans(n_clusters=k, n_init=10)
    kmeans.fit(x)
    
    ax.scatter(x[:, 0], x[:, 1], c=[kmeans.labels_])
    ax.set_title(f"broj_klastera={k}")
    ax.set_xlabel("pca_{1}")
    ax.set_ylabel("pca_{2}")

    sse.append(kmeans.inertia_)
    silh.append(silhouette_score(X, kmeans.labels_))

plt.tight_layout()
plt.show()
```

```python
sses = []
siluete = []
ssesPCA = []
siluetePCA = []
```

```python
for k in range(2,20):
    KSredina = KMeans(n_clusters = k)
    KSredina.fit(X)
    KSredinaPCA = KMeans(n_clusters = k)
    KSredinaPCA.fit(x)

    sses.append(KSredina.inertia_)
    siluete.append(silhouette_score(x, KSredina.labels_))
    ssesPCA.append(KSredinaPCA.inertia_)
    siluetePCA.append(silhouette_score(x, KSredinaPCA.labels_))
```

```python
xs = [*range(2, 20)]

fig,ax = plt.subplots()
ax.plot(xs, sses)
ax.set_xlabel("K #")
ax.set_ylabel("sse")

ax2=ax.twinx()
ax2.plot(xs, siluete, c = 'orange')
ax2.set_ylabel("silueta")
plt.title("silueta i sse originalnih podataka")
plt.show()
```

```python
xs = [*range(2, 20)]

fig,ax = plt.subplots()
ax.plot(xs, ssesPCA)
ax.set_xlabel("K #")
ax.set_ylabel("sse")

ax2=ax.twinx()
ax2.plot(xs, siluetePCA, c = 'orange')
ax2.set_ylabel("silueta")
plt.title("silueta i sse PCE podataka")
plt.show()
```
